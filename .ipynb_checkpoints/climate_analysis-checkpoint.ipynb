# Part 1 - Climate Analysis and Exploration

```python
import os
import numpy as np
import pandas as pd
import matplotlib
import matplotlib.pyplot as plt
from sqlalchemy import create_engine, inspect, func
from sqlalchemy.ext.automap import automap_base
from sqlalchemy.orm import Session
```

# Reflect Tables into SQLAlchemy ORM

```python
db_path = os.path.join('Resources', 'hawaii.sqlite')

engine = create_engine(f'sqlite:///{db_path}')

inspector = inspect(engine)
inspector.get_table_names()

columns_m = inspector.get_columns('measurement')
column_names_m = [column['name'] for column in columns_m]
column_names_m

columns_s = inspector.get_columns('station')
column_names_s = [column['name'] for column in columns_s]
column_names_s
```

```python
db_path = os.path.join('Resources', 'hawaii.sqlite')

engine = create_engine(f'sqlite:///{db_path}')

inspector = inspect(engine)
table_names = inspector.get_table_names()

for table in table_names:
    print(table + ':')
    cols = inspector.get_columns(table)
    col_names = [col['name'] for col in cols]
    print(col_names)
```

```python
Base = automap_base()
Base.prepare(engine, reflect=True)

Measurement = Base.classes.measurement
Station = Base.classes.station

session = Session(bind=engine)
```

# Precipitation Analysis

```python
latest_dt = session.query(func.max(Measurement.date)).first()
latest_dt
```

```python
prcp_12mths = session.query(Measurement.date, Measurement.prcp)\
                        .filter(Measurement.date>'2016-08-23',Measurement.date<='2017-08-23')\
                        .order_by(Measurement.date.asc()).all()
prcp_12mths[0]
```

```python
prcp_12mths_df = pd.DataFrame(prcp_12mths, columns = ['date' , 'prcp']).set_index(['date'])
prcp_12mths_df
```

```python
prcp_12mths_df = prcp_12mths_df.sort_index().dropna()
prcp_12mths_df
```

```python
prcp_12mths_df = prcp_12mths_df.sort_index()

# global matplotlib formatting
matplotlib.rcParams["figure.figsize"] = [50, 10]
matplotlib.rcParams.update({"font.size": 14})

prcp_12mths_df.plot()
plt.title('Precipitation Data of The Last 12 Months')
plt.xlabel('Date')
plt.ylabel('Values of Prcp')
plt.xticks(rotation=90)
plt.ylim([0, 7])
plt.margins(x=0)
plt.grid(True)
plt.savefig('12mths_PrcpData.png')
plt.show()
```

```python
prcp_12mths_df.describe(include='all').round(2)
```

# Station Analysis

```python
num_station = session.query(Station.station).count()
num_station
```

```python
active_station_ls = session.query(Measurement.station, func.count())\
                    .group_by(Measurement.station).order_by(func.count().desc()).all()
active_station_ls
```

```python
active_station_ls[0]
```

```python
USC00519281_info = session.query(func.min(Measurement.tobs),\
                                 func.max(Measurement.tobs),\
                                 func.avg(Measurement.tobs))\
                                .filter(Measurement.station=='USC00519281').all()
USC00519281_info
```

```python
tobs_12mths = session.query(Measurement.date, Measurement.tobs)\
                            .filter(Measurement.station=='USC00519281')\
                            .filter(Measurement.date>'2016-08-23')\
                            .all()

tobs_12mths[0]
```

```python
tobs_12mths_df = pd.DataFrame(tobs_12mths, columns = ['date' , 'tobs']).set_index(['date'])
tobs_12mths_df
x=tobs_12mths_df['tobs']
plt.hist(x, bins=12, edgecolor='black', rwidth=0.5, align='mid')

plt.title('Temperature Data of The Last 12 Months')
plt.xlabel('Values of Tobs')
plt.ylabel('Days(Frequency)')
plt.xlim(59,83)
plt.ylim(0,70)
plt.grid(axis='y')
plt.savefig('12mths_TobData.png')
plt.show()
```
